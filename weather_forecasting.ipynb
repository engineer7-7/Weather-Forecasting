# Forecasting using Linear Regression

```python
# split the dataset
from sklearn.model_selection import train_test_split

X = df.drop('meantemp', axis=1)
y = df['meantemp']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

# Import Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

# create an instance of LinearRegression model
lr = LinearRegression()

# fit the model
lr.fit(X_train, y_train)
```

# Enter new features to predict the new meantemp value

```python
# humitidy
humidity = float(input('Enter Humidity: '))

# wind speed
wind_speed = float(input('Enter Wind Speed: '))

# mean pressure
mean_pressure = float(input('Enter Mean Pressure: '))

# enter year
year = int(input('Enter Year: '))

# enter month
month = int(input('Enter Month: '))

# day
day = int(input('Enter Day: '))

# create an array based on these features
features = np.array([[humidity, wind_speed, mean_pressure, year, month, day]])

# predict the meantemp based on these features

prediction = lr.predict(features)

print(f'Prediction: {prediction}')
```

# MSE (Mean Square Error

```python
from sklearn.metrics import mean_squared_error

total_predictions = lr.predict(X_test)

error = mean_squared_error(y_test, total_predictions)

print(f'MSE: {error}')
```
